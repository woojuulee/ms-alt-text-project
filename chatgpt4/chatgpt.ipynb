# gpt prompt engineering experiments

```python
import os
from openai import AzureOpenAI
from dotenv import load_dotenv

load_dotenv()

import json

file_path = r"\before_gpt_nutri.json"

with open(file_path, 'r') as file:
    data = json.load(file)
data


client = AzureOpenAI(
    # https://learn.microsoft.com/en-us/azure/ai-services/openai/reference#rest-api-versioning
    api_version="2023-07-01-preview",
    # https://learn.microsoft.com/en-us/azure/cognitive-services/openai/how-to/create-resource?pivots=web-portal#create-a-resource
    azure_endpoint="",
    api_key = os.environ.get("AZURE_OPENAI_API_KEY"),

)


completion = client.chat.completions.create(
    model="",  # e.g. gpt-35-instant 
    messages=[
        {
            "role": "user",
            "content": str(data) + 
            """위 딕셔너리의 caption의 value값은 이미지의 그림을 설명하는 것이고,
            ocr의 value값은 이미지의 텍스트 전문이야.이 둘을 전체적으로 종합해서 이미지를 25자 내외의 하나의 문장으로 요약해서 존댓말로 설명해줘.
            근데, ocr의 value값에 식품의 영양성분이나 옷의 상세정보가 들어가있다고 확실하게 판단되면, 
            caption의 value값으로만 한 줄로 요약하고, ocr의 value값은 전문 그대로 답변해줘""",
        },
    ],
)

print(completion.model_dump_json(indent=2))
print(completion.choices[0].message.content)
```

```python
print(completion.choices[0].message.content)
```

# GPT가 중요한 내용 판단 가능한지 알아보는 실험

```python
import os
from openai import AzureOpenAI
from dotenv import load_dotenv
load_dotenv()

import json

file_path = r"C:\Users\User\chatgpt4\fromocr\device.json"

with open(file_path, 'r') as file:
    data = json.load(file)
data


client = AzureOpenAI(
    # https://learn.microsoft.com/en-us/azure/ai-services/openai/reference#rest-api-versioning
    api_version="2023-07-01-preview",
    # https://learn.microsoft.com/en-us/azure/cognitive-services/openai/how-to/create-resource?pivots=web-portal#create-a-resource
    azure_endpoint="",
    api_key = os.environ.get("AZURE_OPENAI_API_KEY"),

)

completion = client.chat.completions.create(
    model="",  # e.g. gpt-35-instant # model="gpt-4-vision-preview",이렇게 해도 답변 나옴.
    messages=[
        {
            "role": "user",
            "content": str(data) + 
            """위 딕셔너리의 caption의 value값은 이미지의 그림을 설명하는 것이고,
            ocr의 value값은 이미지의 텍스트 전문이야.이 둘을 전체적으로 종합해서 이미지를 하나의 문장으로 요약해서 존댓말로 설명해줘.
            근데, ocr의 value값에 사람이 반드시 알아야하는 정보라고 판단되는 내용이 있으면, OCR value값 전체를 설명해줘""",
        },
    ],
)
print(completion.model_dump_json(indent=2))
print(completion.choices[0].message.content)
```
